### Correcting IDs before alignment
This notebook has the objective to unite the CDS from all .gbk while it changes the IDs from each protein. This way we can compare the alignment results! It generate a file called: "all_metagenomes_cds.faa"

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import glob
import os

base_path = "/home/pedro/resultados_antiSMASH/MGYG*"

metagenome_dirs = glob.glob(base_path)

records = []

for path in metagenome_dirs:

    metagenome = os.path.basename(path)

    gbk_files = glob.glob(f"{path}/{metagenome}_*.gbk")

    for gbk in gbk_files:
        for record in SeqIO.parse(gbk, "genbank"):
            for feature in record.features:

                if feature.type == "CDS" and "translation" in feature.qualifiers:

                    locus_tag = feature.qualifiers.get("locus_tag", ["unknown"])[0]
                    protein_seq = feature.qualifiers["translation"][0]

                    cds_id = f"{metagenome}_{locus_tag}"

                    records.append(
                        SeqRecord(
                            Seq(protein_seq),
                            id=cds_id,
                            description=""
                        )
                    )

SeqIO.write(records, "all_metagenomes_cds.faa", "fasta")

print(f"{len(records)} proteínas salvas")
```
