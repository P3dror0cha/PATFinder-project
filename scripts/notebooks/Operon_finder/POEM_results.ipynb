### Uniprot protein alignment analysis
This notebook takes the output from Uniprot API (https://www.ebi.ac.uk/proteins/api/doc/#!/proteins/search). The goal to import the proteins as a fasta file to make the database for further alignment.

```python
import requests
import sys
from tqdm import tqdm

requestURL = (
    "https://rest.uniprot.org/uniprotkb/stream?"
    "query=(keyword:%22Antibiotic%20biosynthesis%22)"
    "&format=fasta"
)

output_file = "uniprot_antibiotic_biosynthesis.faa"

r = requests.get(requestURL, stream=True)

if not r.ok:
    r.raise_for_status()
    sys.exit()

total_size = int(r.headers.get("Content-Length", 0))
block_size = 1024  

with open(output_file, "wb") as f, tqdm(
    total=total_size,
    unit="B",
    unit_scale=True,
    desc="Downloading UniProt"
) as bar:

    for chunk in r.iter_content(block_size):
        if chunk:
            f.write(chunk)
            bar.update(len(chunk))

print("\nDownload finalizado:", output_file)

```
